 we would cut the Vote_Average values and make 4 categories popular, average, below_avg, not_popular to describe it more using catigorise_col() function

```python
def catigorise_col(df, col, labels):

    edges=[df[col].describe()['min'],
           df[col].describe()['25%'],
           df[col].describe()['50%'],
           df[col].describe()['75%'],
           df[col].describe()['max']]

    df[col]=pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df
```

```python
labels=['not_popular','below_avg','average','popular']

catigorise_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```

#### Insights from the Genre Distribution Chart

Drama is the most dominant genre

Drama has the highest number of movies in the dataset, indicating that story-driven content is the most common category.

Comedy, Action, and Thriller are highly popular

These genres have the second-highest counts after Drama, showing a strong focus on entertainment, action-packed, and suspenseful content.

Western, Documentary, and TV Movie genres are least represented

These genres have very few titles compared to others, suggesting lower production volume or audience demand within the dataset.

#### Which has higest votes in the vote_avg columns ?

```python
vote_count =df['Vote_Average'].value_counts()
```

```python
ax= sns.barplot(x=vote_count.index, y= vote_count.values,hue= vote_count.index, palette='viridis')
plt.title('higest votes')

for container in ax.containers:
    ax.bar_label(container,padding=3, fontsize=11)
```

#### Highest Votes by Rating Category
1. Average-rated movies received the highest number of votes
The average category has the highest vote count (6613 votes).
This indicates that most audience engagement is concentrated around moderately rated movies rather than extremely high-rated ones.
    
2. Popular movies are close behind
The popular category has 6520 votes, very close to the average category.
This suggests that well-known movies generally attract large numbers of viewers and ratings.
    
3. Vote counts are fairly balanced across categories
The difference between the lowest (6071) and highest (6613) vote counts is relatively small.
This shows that users actively rate movies across all rating categories, not just the most popular ones.
